```python
# =========================
# LISTS AND BASE DATA
# =========================
racas = ["Humano", "Elfo", "Anao"]
classes = ["Mago", "Arqueiro", "Guerreiro"]
sexo = ["Masculino", "Feminino"]
herois = []

# =========================
# RACE BONUS
# =========================
bonus_raca = {
    "Humano": {"magia": 5},
    "Anao": {"forca": 5},
    "Elfo": {"destreza": 5}
}

# =========================
# ATTACKS BY CLASS
# =========================
ataques_magia = {
    "Firebolt": {"dano": 23, "mp": 10},
    "Thunder": {"dano": 50, "mp": 25},
    "Icelance": {"dano": 30, "mp": 15}
}

ataques_melee = {
    "Slash": {"dano": 18, "mp": 5},
    "Double": {"dano": 30, "mp": 10},
    "Inves": {"dano": 25, "mp": 7}
}

ataques_range = {
    "Double": {"dano": 25, "mp": 12},
    "ArrowRain": {"dano": 45, "mp": 25},
    "Bulleye": {"dano": 29, "mp": 15}
}

ataques_por_classe = {
    "Mago": ataques_magia,
    "Guerreiro": ataques_melee,
    "Arqueiro": ataques_range
}

# =========================
# CLASS BONUS
# =========================
bonus_classe = {
    "Mago": {"magia": 3},
    "Guerreiro": {"resistencia": 3},
    "Arqueiro": {"destreza": 3}
}

# =========================
# MENU FUNCTION
# =========================
def menu_opcoes(lista, titulo):
    print(f"\n{titulo}")
    for i, item in enumerate(lista):
        print(f"{i + 1}. {item}")
    while True:
        escolha = input("N칰mero: ")
        if escolha.isdigit():
            escolha = int(escolha)
            if 1 <= escolha <= len(lista):
                return lista[escolha - 1]
        print("Op칞칚o inv치lida.")

# =========================
# HERO CREATION
# =========================
while True:
    nome = input("\nDigite o nome do personagem: ")
    raca_escolhida = menu_opcoes(racas, "Escolha sua ra칞a:")
    classe_escolhida = menu_opcoes(classes, "Escolha sua classe:")
    sexo_escolhido = menu_opcoes(sexo, "Escolha o sexo do personagem:")

    atributos = {
        "destreza": 10,
        "magia": 10,
        "forca": 10,
        "resistencia": 10,
        "mp": 100,
        "hp": 200
    }

    for attr, bonus in bonus_raca[raca_escolhida].items():
        atributos[attr] += bonus

    for attr, bonus in bonus_classe[classe_escolhida].items():
        atributos[attr] += bonus

    heroi = {
        "nome": nome,
        "raca": raca_escolhida,
        "classe": classe_escolhida,
        "sexo": sexo_escolhido,
        "status": atributos,
        "ataques": ataques_por_classe[classe_escolhida],
        "xp": 0,
        "nivel": 1
    }

    herois.append(heroi)

    print(f"\n{nome} criado com sucesso!")
    for k, v in atributos.items():
        print(f"{k.capitalize()}: {v}")

    if input("\nCriar outro personagem? (s/n): ").lower() != "s":
        
        break

# =========================
# HERO SELECTION
# =========================
for i, h in enumerate(herois):
    print(f"{i+1}. {h['nome']} ({h['classe']})")

h1 = herois[int(input("Escolha o her칩i 1: ")) - 1]
h2 = herois[int(input("Escolha o her칩i 2: ")) - 1]

# =========================
# COMBAT SYSTEM (QUEST 4)
# =========================
turno = 0
while h1["status"]["hp"] > 0 and h2["status"]["hp"] > 0:
    atacante = h1 if turno % 2 == 0 else h2
    defensor = h2 if turno % 2 == 0 else h1
    is_player = atacante == h1

    print(f"\nTurno {turno + 1} - {atacante['nome']} ataca")
    print(f"MP: {atacante['status']['mp']} | HP inimigo: {defensor['status']['hp']}")

    ataques = atacante["ataques"]
    nomes_ataques = list(ataques.keys())

    # ATTACK SELECTION
    if not is_player:
        ataques_validos = [
            atk for atk in nomes_ataques
            if ataques[atk]["mp"] <= atacante["status"]["mp"]
        ]

        if not ataques_validos:
            print(f"{atacante['nome']} perdeu o turno (MP insuficiente).")
            turno += 1
            continue

        ataque_escolhido = ataques_validos[0]

    else:
        for i, atk in enumerate(nomes_ataques):
            print(f"{i+1}. {atk} (Dano {ataques[atk]['dano']}, MP {ataques[atk]['mp']})")

        while True:
            escolha = input("N칰mero do ataque: ")
            if escolha.isdigit():
                escolha = int(escolha)
                if 1 <= escolha <= len(nomes_ataques):
                    ataque_escolhido = nomes_ataques[escolha - 1]
                    if ataques[ataque_escolhido]["mp"] <= atacante["status"]["mp"]:
                        break
                    else:
                        print("MP insuficiente.")
            print("Op칞칚o inv치lida.")

    # DAMAGE CALCULATION
    ataque = ataques[ataque_escolhido]
    atacante["status"]["mp"] -= ataque["mp"]

    if atacante["classe"] == "Mago":
        bonus = atacante["status"]["magia"]
    elif atacante["classe"] == "Guerreiro":
        bonus = atacante["status"]["forca"]
    else:
        bonus = atacante["status"]["destreza"]

    dano = ataque["dano"] + bonus
    defensor["status"]["hp"] -= dano

    print(f"{atacante['nome']} usou {ataque_escolhido} e causou {dano} de dano!")

    turno += 1

# =========================
# RESULT
# =========================
vencedor = h1 if h1["status"]["hp"] > 0 else h2
print(f"\n游끥 {vencedor['nome']} venceu o combate!")

# =========================
# XP AND LEVEL UP
# =========================
vencedor["xp"] += 10
while vencedor["xp"] >= 20 * vencedor["nivel"]:
    vencedor["xp"] -= 20 * vencedor["nivel"]
    vencedor["nivel"] += 1
    print(f"{vencedor['nome']} subiu para o n칤vel {vencedor['nivel']}!")
```
